```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import OneHotEncoder

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, FunctionTransformer
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor, VotingRegressor, StackingRegressor
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
df_train = pd.read_csv("/content/train (1).csv")
df_test = pd.read_csv("/content/test (1).csv")
```

```python
df_train.drop(columns=['id'],inplace=True)
```

```python
df_train.head()
```

```python
col = "Age"
df_log = df_train.copy()
df_log[col + "_log"] = np.log1p(df_log[col]+1)
```

```python
df_log.head()
```

```python
df_log.drop(columns=['Age'],inplace=True)
```

```python
df_log.head()
```

```python
X = df_log.drop('Age_log', axis=1)
y = df_log['Age_log']
```

```python
#X = df_train.drop('Age', axis=1)
#y = df_train['Age']
```

```python
numerical_cols = X.select_dtypes(include=np.number).columns
categorical_cols = X.select_dtypes(include=object).columns
```

```python
print("Numerical columns:", numerical_cols)
print("Categorical columns:", categorical_cols)
```

```python
!pip install feature_engine
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import FunctionTransformer
import numpy as np

num_pipe = Pipeline(
    [
        ('imputer', SimpleImputer(strategy='median')),
         ('log', FunctionTransformer(func=np.log1p, validate=False)),
        ('scaler', StandardScaler())
    ]
)
```

```python
cat_pipe = Pipeline(
    [
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('encoder', OneHotEncoder(handle_unknown='ignore'))
    ]
)
```

```python
combine_pipe = ColumnTransformer(
    [
        ('num', num_pipe, numerical_cols),
        ('cat', cat_pipe, categorical_cols)
    ]
)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
#base learner

lr = LinearRegression()
dt = DecisionTreeRegressor()
rf = RandomForestRegressor()
gb = GradientBoostingRegressor()
ada = AdaBoostRegressor()
xgb = XGBRegressor()
lgb = LGBMRegressor()
cat = CatBoostRegressor()
```

```python
#base learner

lrn = LinearRegression()
dtr = DecisionTreeRegressor()
rfr = RandomForestRegressor()
gbr = GradientBoostingRegressor()
adar = AdaBoostRegressor()
xgbr = XGBRegressor()
lgbr = LGBMRegressor()
catr = CatBoostRegressor()
```

```python
# used voting

voting_rg = VotingRegressor(
    estimators=[
        ('lr', lr),
        ('rf', rf),
        ('gb', gb)
    ]
)
```

```python
stacking_rg = StackingRegressor(
    estimators=[
    ('lr', lr),
    ('dt', dt),
    ('rf', rf),
    ('gb', gb),
    ('xgb', xgb)
    ]
)
```

```python
model_train = {
    'Linear Regression': lr,
    'Decision Tree': dt,
    'Random Forest': rf,
    'Gradient Boost': gb,
    'AdaBoost': ada,
    'Voting Regressor': voting_rg,
    'Stacking Regressor': stacking_rg

}
```

```python
import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# Store results
results = []

# Loop through all regression models
for name, model in model_train.items():
    # Create pipeline: combine preprocessing pipeline + model
    pipe = Pipeline([
        ('preprocessing', combine_pipe),
        ('regressor', model)
    ])

    # Fit the model
    pipe.fit(X_train, y_train)

    # Predict on test set
    y_pred = pipe.predict(X_test)

    # Evaluate
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)


    # Store results in dictionary
    results.append({
        'Model Name': name,
        'R2 Score': r2,
        'MAE': mae,
        'MSE': mse,
        'RMSE': rmse
    })

# Convert results to DataFrame and sort by R2 Score
result_df = pd.DataFrame(results).sort_values(by='R2 Score', ascending=False).reset_index(drop=True)

print(result_df)
```

```python
CatBR = CatBoostRegressor()

pipe_ct = Pipeline([
        ('preprocessing', combine_pipe),
        ('regressor', CatBR)
    ])
pipe_ct.fit(X_train, y_train)
y_pred = pipe_ct.predict(X_test)
print(r2_score(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
print(np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
Xgb = XGBRegressor()
xgb_pipe = Pipeline([
        ('preprocessing', combine_pipe),
        ('regressor', Xgb)
    ])

xgb_pipe.fit(X_train, y_train)
y_pred = xgb_pipe.predict(X_test)
print(r2_score(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
print(np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
lgbm = LGBMRegressor()

lgbm_pipe = Pipeline(
    [
        ('preprocessing', combine_pipe),
        ('regressor', lgbm)
]
)

lgbm_pipe.fit(X_train, y_train)
y_pred = lgbm_pipe.predict(X_test)
print(r2_score(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
print(np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
best_model_name = result_df.iloc[0]['Model Name']
best_model_obj = model_train[best_model_name]

# 2 Create final pipeline
final_pipe = Pipeline([
    ('Combine_pipe', combine_pipe),  # your numeric + categorical preprocessing
    ('regressor', best_model_obj)
])

# 3 Fit the final model
final_pipe.fit(X_train, y_train)

# 4 Predict on test set
y_final_pred = final_pipe.predict(X_test)

# 5 Evaluate metrics
r2 = r2_score(y_test, y_final_pred)
mae = mean_absolute_error(y_test, y_final_pred)
mse = mean_squared_error(y_test, y_final_pred)
rmse = np.sqrt(mse)

print(f"Best Model: {best_model_name}")
print(f"R Score: {r2:.4f}")
print(f"Mean_absolute_error: {mae:.4f}")
print(f"Mean_squared_error: {mse:.4f}")
print(f"RMSE: {rmse:.4f}")

# 6 Optional: plot predicted vs actual
plt.figure(figsize=(8,6))
sns.scatterplot(x=y_test, y=y_final_pred)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title(f'{best_model_name}: Actual vs Predicted')
plt.show()
```

```python
# Crossn Validation
rf_cross_pipe = Pipeline(
    [
        ('preprocessing', combine_pipe),
        ('regressor', RandomForestRegressor(n_estimators=[500,1000],max_depth=[8,12],min_samples_leaf=[2,5]))
    ]
)
```

```python
# Fix the RandomForestRegressor definition for cross_val_score
# The previous rf_cross_pipe was defined with lists for n_estimators, max_depth, and min_samples_leaf,
# which are not valid for direct instantiation of RandomForestRegressor in cross_val_score.
# We will use the first value from each list as a default for this cross-validation run.
fixed_rf_cross_pipe = Pipeline(
    [
        ('preprocessing', combine_pipe),
        ('regressor', RandomForestRegressor(n_estimators=500, max_depth=8, min_samples_leaf=2, random_state=42)) # Added random_state for reproducibility
    ]
)

cross_val_scores_rf = cross_val_score(fixed_rf_cross_pipe, X_train, y_train, scoring='r2', cv=5)

print("Cross Value Score 5 cv Randomforest", cross_val_scores_rf)
```
